```python
import pickle
import pandas as pd
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split , cross_val_score, cross_val_predict, GridSearchCV, RandomizedSearchCV
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn import metrics
```

```python
with open('model_df.pickle','rb') as read_df:
    df = pickle.load(read_df)
```

```python
df.head()
```

```python
X = df.drop(columns=['cleaned_current_value'])
Y = df['cleaned_current_value']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=.2, random_state=88)
```

```python
num_features = ['released','180 days']
cat_features = ['brand_series']
```

```python
coltransf = make_column_transformer((StandardScaler(),num_features),(OneHotEncoder(handle_unknown='ignore'),cat_features))
```

```python
linreg = LinearRegression()
```

```python
pipeline = make_pipeline(coltransf,linreg)
```

```python
linreg = cross_val_score(pipeline, X_train, y_train)
```
