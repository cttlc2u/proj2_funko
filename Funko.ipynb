```python
subv_links = []
driver = webdriver.Chrome(chromedriver)
for i in subvariants['Link']:
    url = ppg + i
    driver.get(url)
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    variants_page = soup.find_all('a', class_='catalog-item-name ng-binding')
    for link in variants_page:
        name = link.text
        name = name.strip()
        page = link.get("href")
        page = page.replace('/marketplaces/poppriceguide/catalog_items/','')
        subv_links.append([name,page])
        time.sleep(2)
with open('subv_links.pickle', 'wb') as subv_links_pickle:
    pickle.dump(subv_links, subv_links_pickle)
```
