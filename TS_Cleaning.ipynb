```python
values_dict = {}
for idx, time in enumerate(df['values']):
    # number of sales print(time[1::6])
    # Avg. print(time[2::6])
    # Low. print(time[3::6])
    # Median print(time[4::6])
    # High print(time[5::6])
    hist_price_dict = dict(zip(time[::6],time[2::6]))
    values_dict[df.index[idx]] = hist_price_dict
```
