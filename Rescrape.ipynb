```python
directory = 'HTMLs'
poppages = []
for filename in os.listdir(directory): 
    f = os.path.join(directory, filename) 
    if os.path.isfile(f): 
        poppages.append(f)
```

```python
poppages
```

```python
pop_df_dict = {}

for file in poppages:
    single_pop_dict = {}
    with open(file) as f:
        soup = BeautifulSoup(f.read())
        if soup.find('h1', class_='item-name'):
            single_pop_dict['name'] = soup.find('h1', class_='item-name').text.strip()
        if soup.find(string=re.compile('Alternative Name: ')):
            single_pop_dict['alt_name'] = soup.find(string=re.compile('Alternative Name: ')).strip().replace('Alternative Name: ','')
        detailslist = []
        for item in soup.find_all('b'):
            if item.text.strip()[0] != '$' and item.text.strip()!='Released:' and item.text.strip() != 'Production Status:':
                detailslist.append(item.text.strip())
        detailsvalues = []
        for item in soup.find_all('a', class_='ng-binding',href=""):
            detailsvalues.append(item.text.strip())
        if len(detailslist)==len(detailsvalues[:-1]):
            detailsdict = dict(zip(detailslist,detailsvalues))
            single_pop_dict['details'] = detailsdict
        if soup.find('div',class_='col-md-6 spaced-field ng-scope'):
            if soup.find('div',class_='col-md-6 spaced-field ng-scope').find('b'):
                single_pop_dict['released'] = soup.find('div',class_='col-md-6 spaced-field ng-scope').find(class_='ng-binding').text.strip()[:4]
        brand_series_list = []
        for item in soup.find_all('div', class_='ng-binding ng-scope'):
            brand_series_list.append(item.text.strip())
        single_pop_dict['brand_series'] = brand_series_list
        value_list = []
        if soup.find_all('td',class_='ng-binding'):
            for item in soup.find_all('td',class_='ng-binding'):
                value_list.append(item.text.strip())
        single_pop_dict['values'] = value_list
        if soup.find('div', class_='value ng-binding'):
            single_pop_dict['current_value'] = soup.find('div', class_='value ng-binding').text.strip()
        pop_df_dict[file] = single_pop_dict
```

```python
with open('ppg_dict.pickle', 'wb') as ppg_dict_pickle:
    pickle.dump(pop_df_dict, ppg_dict_pickle)
```
