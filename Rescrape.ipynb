```python
chromedriver = "/Applications/chromedriver"
os.environ["webdriver.chrome.driver"] = chromedriver
ppg = 'https://www.hobbydb.com/marketplaces/poppriceguide/catalog_items/'
```

```python
driver = webdriver.Chrome(chromedriver)
for idx, link in enumerate(df['Link'][:]):
    url = ppg + link
    driver.get(url)
    filename = "ppg_" + str(idx) + ".txt"
    with open('HTMLs/'+filename, "w") as f:
        f.write(driver.page_source)
    time.sleep(2+1*random.random())
```
