```python
ppg = 'https://www.hobbydb.com/marketplaces/poppriceguide/catalog_items/'
```

```python
ppg_details = defaultdict(str)
driver = webdriver.Chrome(chromedriver)
for link in df['Link'][:]:
    url = ppg + link
    driver.get(url)
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    price = soup.find('div', class_='value ng-binding')
    if price:
        price = price.text.strip()
    fields = soup.find_all('b')
    fields_list = []
    for field in fields:
        fields_list.append(field.text.strip())
    details_list = []
    details = soup.find_all('a', class_='ng-binding',href="")
    for detail in details:
        details_list.append(detail.text.strip())
    prod_status = soup.find_all('b', class_='')
    if prod_status:
        prod_status_list = []
        for yes in prod_status:
            prod_status_list.append(yes.text.strip())
        exclusive = soup.find_all('li')
        for exc in exclusive:
            prod_status_list.append(exc.text.strip())
    time.sleep(2)
    ppg_details[link] = [price,fields_list,details_list,prod_status_list]

#with open('ppg.pickle', 'wb') as ppg_pickle:
    #pickle.dump(ppg_details, ppg_pickle)
```
